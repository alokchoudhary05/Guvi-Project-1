# Movie Ratings Analysis

```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For better visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

## Objective

We'll explore a dataset of 1000 movies and perform the following:

- Create a table showing Movie Name and its Rating  
- Bar chart: Average rating per genre  
- Pie chart: Movie count by genre  
- Show top 5 highest-rated movies  

```python
# Step 2: Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/movie_ratings_dataset.csv")
df.head()
```

```python
# Step 3: Table - Movie name and rating
df_table = df[["Title", "Rating"]]
df_table.head(10)
```

```python
# Step 4: Bar plot - Average rating per genre
avg_rating = df.groupby("Genre")["Rating"].mean().sort_values(ascending=False)

sns.barplot(x=avg_rating.index, y=avg_rating.values, palette="viridis")
plt.title("Average Rating per Genre", fontsize=16)
plt.xlabel("Genre")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Step 5: Pie chart - Movie count by genre
genre_counts = df["Genre"].value_counts()

plt.pie(genre_counts.values, labels=genre_counts.index, autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Movie Count by Genre", fontsize=16)
plt.axis("equal")
plt.show()
```

```python
# Step 6: Top 5 highest-rated movies
top5 = df.sort_values(by="Rating", ascending=False).head(5)
top5[["Title", "Rating", "Genre"]]
```

##  Summary

- We visualized the distribution of ratings and genre popularity.
- Found out which genres perform best on average.
- Listed the top 5 movies based on rating.
